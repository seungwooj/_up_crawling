```python
import requests
import time
from bs4 import BeautifulSoup
```

```python
def comments():
    df = pd.DataFrame(columns=["age","comments"])
    response = requests.get("http://widgets4.cre.ma/bodyluv.kr/products/reviews?product_code=52&iframe_id=crema-product-reviews-2&app=0&parent_url=http%3A%2F%2Fwww.bodyluv.kr%2Fproduct%2F%25EB%25A7%2588%25EC%2595%25BD%25EB%25A7%25A4%25ED%258A%25B8%25EB%25A6%25AC%25EC%258A%25A4%2F52%2Fcategory%2F25%2Fdisplay%2F1%2F&iframe=1")             # response : string
    dom = BeautifulSoup(response.content, "html.parser")    # response를 beautifulsoup을 활용해서 parsing
    keywords = dom.select("products_reviews_list_review__message")
    for keyword in keywords:
        df.loc[len(df)] = {
            "age":keyword.select_one("review_option__content").text,
            "comments":keyword.select_one("products_reviews_list_review__message_content").text,
        }
    return df
```
